```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/heart.csv")
df.head()
```

```python
oneHotFeatures = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
df = pd.get_dummies(data = df, 
                    prefix = oneHotFeatures,
                    columns = oneHotFeatures)
df.head()
```

```python
features = [x for x in df.columns if x not in 'HeartDisease']
```

```python
xTrain, xValid, yTrain, yValid = train_test_split(df[features], df['HeartDisease'], train_size = 0.8)
```

```python
minSamplesSplits = [2,10, 30, 50, 100, 200, 300, 700]

accuracysByTrain = []
accuracysByValid = []

for val in minSamplesSplits:
    model = DecisionTreeClassifier(min_samples_split = val).fit(xTrain, yTrain)
    
    trainAccuracy = accuracy_score(model.predict(xTrain), yTrain)
    validAccuracy = accuracy_score(model.predict(xValid), yValid)
    accuracysByTrain.append(trainAccuracy)
    accuracysByValid.append(validAccuracy)

plt.title('Train x Validation metrics')
plt.xlabel('min_samples_split')
plt.ylabel('accuracy')
plt.plot(accuracysByTrain)
plt.plot(accuracysByValid)
plt.legend(['Train','Validation'])
```

```python
maxDepths = [1, 2, 3, 4, 8, 16, 32, 64, None]

accuracysByTrain = []
accuracysByValid = []

for val in maxDepths:
    model = DecisionTreeClassifier(max_depth = val).fit(xTrain, yTrain)
    
    trainAccuracy = accuracy_score(model.predict(xTrain), yTrain)
    validAccuracy = accuracy_score(model.predict(xValid), yValid)
    accuracysByTrain.append(trainAccuracy)
    accuracysByValid.append(validAccuracy)

plt.title('Train x Validation metrics')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.plot(accuracysByTrain)
plt.plot(accuracysByValid)
plt.legend(['Train','Validation'])
```

```python
model = DecisionTreeClassifier()
paramGrid = {
    'min_samples_split': minSamplesSplits,
    'max_depth': maxDepths,
}

gridSearch = GridSearchCV(estimator = model, param_grid = paramGrid)
gridSearch.fit(xTrain, yTrain)

print(f"Best parameters found: {gridSearch.best_params_}")

model = gridSearch.best_estimator_
print(f"Metrics train:\n\tAccuracy score: {accuracy_score(model.predict(xTrain), yTrain): .4f}")
print(f"Metrics validation:\n\tAccuracy score: {accuracy_score(model.predict(xValid), yValid): .4f}")
```

```python
model = RandomForestClassifier()
paramGrid = {
    'n_estimators': [1, 10, 20, 50, 100],
    'min_samples_split': minSamplesSplits,
    'max_depth': maxDepths,
}

gridSearch = GridSearchCV(estimator = model, param_grid = paramGrid)
gridSearch.fit(xTrain, yTrain)

print(f"Best parameters found: {gridSearch.best_params_}")

model = gridSearch.best_estimator_
print(f"Metrics train:\n\tAccuracy score: {accuracy_score(model.predict(xTrain), yTrain): .4f}")
print(f"Metrics validation:\n\tAccuracy score: {accuracy_score(model.predict(xValid), yValid): .4f}")
```

```python
n = int(len(xTrain) * 0.8)
xTrainSet = xTrain[:n]
xEval = xTrain[n:]
yTrainSet = yTrain[:n]
yEval = yTrain[n:]

model = XGBClassifier(n_estimators = 500, learning_rate = 0.1, early_stopping_rounds = 10)
model.fit(xTrain, yTrain, eval_set = [(xEval, yEval)], verbose = False)

print(f"Metrics train:\n\tAccuracy score: {accuracy_score(model.predict(xTrain), yTrain): .4f}")
print(f"Metrics validation:\n\tAccuracy score: {accuracy_score(model.predict(xValid), yValid): .4f}")
```
