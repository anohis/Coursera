```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loadData(path):
    data = np.loadtxt(path, delimiter = ',')
    x = data[:,:2]
    y = data[:,2]
    return x, y
```

```python
def plotData(x, y):
    pos = y == 1
    neg = y == 0
    plt.plot(x[pos, 0], x[pos, 1], 'k+', label = 'Admitted')
    plt.plot(x[neg, 0], x[neg, 1], 'yo', label = 'Not admitted')
```

```python
x, y = loadData("data/ex2data1.txt")
plotData(x, y)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def model(x, w, b):
    z = np.matmul(x, w) + b
    return sigmoid(z)

def computeCost(x, y, w, b, lambda_):
    pred = model(x, w, b)
    loss = - (y * np.log(pred) + (1 - y) * np.log(1 - pred))
    cost = np.mean(loss) / 2 + (lambda_ / (2 * len(x))) * np.sum(w ** 2)
    return cost

def computeGradient(x, y, w, b, lambda_):
    pred = model(x, w, b)
    err = pred - y
    err = np.expand_dims(err, axis = 1)
    gradW = np.mean(err * x, axis = 0) + lambda_ / len(x) * w
    gradB = np.mean(err)
    return gradW, gradB

def gradientDescent(x, y, w, b, lambda_, alpha, iterCount):
    costs = []
    for i in range(iterCount):
        gradW, gradB = computeGradient(x, y, w, b, lambda_)
        w = w - alpha * gradW
        b = b - alpha * gradB
        cost = computeCost(x, y, w, b, lambda_)
        costs.append(cost)

        if i % (iterCount / 10) == 0:
            print(f"iter: {i}, cost: {cost:0.5f}")
    return w, b, costs
```

```python
def plotLinearBoundary(x, y, w, b):
    plotData(x, y)

    x = np.array([min(x[:, 0]), max(x[:, 0])])
    y = (-1. / w[1]) * (w[0] * x + b)
    plt. plot(x, y, c = "blue")
```

```python
np.random.seed(1)
w = 0.01 * (np.random.rand(x.shape[1]) - 0.5)
b = -8
lambda_ = 0.01
alpha = 0.001
iterCount = 10000

w, b, costs = gradientDescent(x, y, w, b, lambda_, alpha, iterCount)

plotLinearBoundary(x, y, w, b)
```

### Feature mapping

將兩個特徵 $x_1$ 和 $x_2$ 進行6接多項式.

$$\mathrm{map\_feature}(x) = 
\left[\begin{array}{c}
x_1\\
x_2\\
x_1^2\\
x_1 x_2\\
x_2^2\\
x_1^3\\
\vdots\\
x_1 x_2^5\\
x_2^6\end{array}\right]$$

兩個特徵最後會變成27維度向量

```python
def mapFeature(x1, x2):
    x1 = np.atleast_1d(x1)
    x2 = np.atleast_1d(x2)
    degree = 6
    out = []
    for i in range(1, degree + 1):
        for j in range(i + 1):
            out.append((x1 ** (i - j) * (x2 ** j)))
    return np.stack(out, axis = 1)
```

```python
def plotPolynomialBoundary(x, y, w, b):
    plotData(x, y)

    x1 = np.linspace(-1, 1.5, 50)
    x2 = np.linspace(-1, 1.5, 50)
    predMatrix = np.zeros((len(x1), len(x2)))
    for i in range(len(x1)):
        for j in range(len(x2)):
            x = mapFeature(x1[i], x2[j])
            predMatrix[i][j] = model(x, w, b)[0]
    predMatrix = predMatrix.T
    plt.contour(x1, x2, predMatrix, levels = [0.5], colors = "g")
```

```python
mappedX = mapFeature(x[:, 0], x[:, 1])

np.random.seed(1)
w = np.random.rand(mappedX.shape[1]) - 0.5
b = 1
lambda_ = 0.01
alpha = 0.01
iterCount = 10000

w, b, costs = gradientDescent(mappedX, y, w, b, lambda_, alpha, iterCount)

plotPolynomialBoundary(x, y, w, b)
```
