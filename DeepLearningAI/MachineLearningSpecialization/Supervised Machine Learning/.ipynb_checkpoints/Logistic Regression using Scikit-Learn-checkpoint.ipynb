```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
x = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(x, y)
pred = model.predict(x)

yTrue = y == 1
yFalse = y == 0
predTrue = pred >= 0.5
predFalse = pred < 0.5

fig, ax = plt.subplots(1, 2)
ax[0].scatter(x[yTrue, 0], x[yTrue, 1], marker = "o", color = "blue")
ax[0].scatter(x[yFalse, 0], x[yFalse, 1], marker = "x", color = "red")
ax[0].set_title("actual")

ax[1].scatter(x[predTrue, 0], x[predTrue, 1], marker = "o", color = "blue")
ax[1].scatter(x[predFalse, 0], x[predFalse, 1], marker = "x", color = "red")
ax[1].set_title("pred")
```
